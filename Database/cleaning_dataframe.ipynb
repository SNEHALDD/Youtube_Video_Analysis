```python
#bin video length?
joined_df=pd.read_csv("ready_for_sql/joined_data.csv")
joined_df['video_length'].unique
```

```python
joined_df["video_length"].describe()
```

```python
joined_df.dtypes
```

```python
#replace the : with . so maybe it will bin?
joined_df['new_length']=joined_df['video_length'].replace(":",".")
joined_df
```

```python
#bin channel_view_count
joined_df["channel_view_count"].describe()
```

```python
channel_views_labels=['0-1000 views','1,000-100,000 views','100,000-1 mil views','1 mil-5 mil views','5 mil- 100 mil','100 mil- 500 mil views','500 mil- 1 billion views','1 bil- 100 billion views', 'over 100 billion views']
channel_views_bins=[0,1000,100000,1000000,5000000,100000000,500000000,1000000000,100000000000,150000000000]

joined_df["channel_views_binned"]=pd.cut(joined_df["channel_view_count"], channel_views_bins, labels=channel_views_labels)
joined_df
```

```python
#bin subscriber count
joined_df["subscriber_count"].describe()
```

```python
subscriber_labels=['under 100 subs','100-1,000 subs','1,000-5,000 subs','5,000-10,000 subs','10,000-500,000 subs','500,000- 1 mil subs',' 1-5 mil subs','5-15 mil subs','15-50 mil subs','50-100 mil subs','over 100 mil subs']
subscriber_bins=[0,100,1000,5000,10000,500000,1000000,5000000,15000000,50000000,100000000,150000000]

joined_df["subscribers_binned"]=pd.cut(joined_df["subscriber_count"], subscriber_bins, labels=subscriber_labels)
joined_df

```

```python
joined_df['channel_video_count'].describe()
```

```python
video_count_labels=['less than 100 videos','100-500 videos','500-1500 videos','1500-5000 videos','5,000-10,000 videos','greater than 10,000 videos']
video_count_bins=[0,100,500,1500,5000,10000,246686]

joined_df["video_count_binned"]=pd.cut(joined_df["channel_video_count"], video_count_bins, labels=video_count_labels)
joined_df
```

```python
#day of week video published
joined_df["datetime"]=pd.to_datetime(joined_df['published_at'])
joined_df['day_of_week_published']=joined_df['datetime'].dt.day_name()
joined_df
```

```python
#bin like count
joined_df['like_count'].describe()
```

```python
like_count_labels=['less than 50 likes','50-1000 likes','1,000-5,000 likes','5,000-25,000 likes','25,000-500,000','over 1 million likes','over 10 million likes']
like_count_bins=[0,50,1000,5000,25000,500000,1000000,15000000]

joined_df["like_count_binned"]=pd.cut(joined_df["like_count"], like_count_bins, labels=like_count_labels)
joined_df
```

```python
#comment count describe
joined_df['comment_count'].describe()
```

```python
#bin comment count
comment_labels=['no comments','less than 100 comments','100-500 comments','500-1000 comments','1,000-10,000 comments','10,000-100,000 comments', 'over 100,000 comments']
comment_bins=[0,1,100,500,1000,10000,100000,800000]

joined_df["comment_binned"]=pd.cut(joined_df["comment_count"], comment_bins, labels=comment_labels)

joined_df['comment_binned']=joined_df['comment_binned'].fillna('no comments')

joined_df
```

```python
#bin video view count
joined_df['view_count'].describe()
```

```python
view_labels=['less than 100 views','less than 1000 views','1,000-10,000 views','10,000-50,000 views','50,000-500,000','500,000- 1 million views','over 1 million views','over 5 million views','over 1 billion views']
view_bins=[0,100,1000,10000,50000,500000,1000000,5000000,1000000000,3000000000]

joined_df["video_views_binned"]=pd.cut(joined_df["view_count"], view_bins, labels=view_labels)
joined_df
```

```python
joined_df=joined_df.drop(['new_length','datetime'],axis=1)
joined_df
```
